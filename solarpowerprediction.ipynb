```python
# plot distribution of power
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30,kde=True) # kde is kernel density estimation
plt.title('Distribution of Power Generated (kW)')
plt.xlabel('Power Generated (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
#scatter-plot with target feature
plt.figure(figsize=(15,20))
for i, column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column],df["generated_power_kw"])
    plt.title(f'df{column} vs Generated_power (kw)')
    plt.xlabel(column)
    plt.ylabel("Generated power (kw)")
plt.tight_layout()
plt.show()
```
